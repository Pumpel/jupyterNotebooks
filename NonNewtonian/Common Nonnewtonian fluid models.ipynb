# Common Nonnewtonian Fluid Models

The goal of this document is to provide a comprehensive list of commonly used nonnewtonian fluid flow models
and  visualize their properties if applicable. The models are taken from the talk: "Some frequently used models for non-Newtonian
fluids" Josef Malek.
Here, the stress tensor $\textbf{T}$ is defined by:

$$ \textbf{T} = -p\textbf{I} + 2 \mu(\textbf{T},\textbf{D})\textbf{D} $$


# 1 Models With Variable Viscosity

## 1.1 Models With Shear Velocity Dependend Viscosity

- **Ostwald–de Waele, Power Law Fluid**
 - Wolfgang Ostwald. Uber die Geschwindigkeitsfunktion der ¨
Viskosit¨at disperser Systeme. I. Colloid Polym. Sci., 36:99–117,
1925
A. de Waele. Viscometry and plastometry. J. Oil Colour Chem.
Assoc., 6:33–69, 1923
 - $\mu(\textbf{D}) = \mu_0 \left| \textbf{D} \right|^{n-1}$
 - n < 1, pseudo-plastic
 - n = 1, newtonian
 - n > 1, dilatant
 - Fits experimental data for: ball point pen ink, molten chocolate, 
aqueous dispersion of polymer latex spheres
- **Carreau / Carreau–Yasuda**
 - Pierre J. Carreau. Rheological equations from molecular network
theories. J. Rheol., 16(1):99–127, 1972
Kenji Yasuda. Investigation of the analogies between viscometric
and linear viscoelastic properties of polystyrene fluids. PhD thesis,
Massachusetts Institute of Technology. Dept. of Chemical
Engineering., 1979
 - $ \mu(\textbf{D}) = \mu_\infty + \frac{\mu_0 - \mu_\infty}{(1 + \alpha \left| \textbf{D} \right|^2)^{n/2}}$
 - $ \mu_\infty + (\mu_0 - \mu_\infty) (1 + \alpha \left| \textbf{D} \right|^2)^{a})^{ \frac{n-1}{a}} $ 
 - Fits molten polystyrene
 - At low shear rate Carreau fluid behaves as a Newtonian fluid and at high shear rate as a power-law fluid.
- **Eyring**
 - Henry Eyring. Viscosity, plasticity, and diffusion as examples of
absolute reaction rates. J. Chem. Phys., 4(4):283–291, 1936
Francis Ree, Taikyue Ree, and Henry Eyring. Relaxation theory of
transport problems in condensed systems. Ind. Eng. Chem.,
50(7):1036–1040, 1958
 -napalm (coprecipitated aluminum salts
of naphthenic and palmitic acids; jellied gasoline)
 - $ \mu(\textbf{D}) = \mu_\infty + (\mu_0 - \mu_\infty) \frac{arcsinh(\alpha \left| \textbf{D} \right|)}{\alpha \left| \textbf{D} \right|}$
 - Fits experimental data for: napalm (coprecipitated aluminum salts
of naphthenic and palmitic acids; jellied gasoline)
- **Cross**
 - Malcolm M. Cross. Rheology of non-newtonian fluids: A new flow
equation for pseudoplastic systems. J. Colloid Sci., 20(5):417–437,
1965
 - $ \mu(\textbf{D}) = \mu_\infty + \frac{\mu_0 - \mu_\infty}{1 + \alpha \left| \textbf{D} \right|^n} $
 - Fits experimental data for: aqueous polyvinyl acetate dispersion,
aqueous limestone suspension
- **Sisko**
 - C. Barus. Isotherms, isopiestics and isometrics relative to viscosity.
Amer. J. Sci., 45:87–96, 1893
 - $ \mu(\textbf{D}) = \mu_\infty + \alpha \left| \textbf{D} \right|^n $
 - Fits experimental data for: lubricating greases

## 1.2 Model With Pressure Dependent Viscosity
- **Barus**
 - C. Barus. Isotherms, isopiestics and isometrics relative to viscosity.
Amer. J. Sci., 45:87–96, 1893
 - $ \mu(\textbf{T}) = \mu_{ref} + e^{\beta(p- p_{ref})} $
 - Fits experimental data for: mineral oils1, organic liquids

## 1.3 Models With Stress Dependend Viscosity
- **Ellis**
 - Seikichi Matsuhisa and R. Byron Bird. Analytical and numerical
solutions for laminar flow of the non-Newtonian Ellis fluid. AIChE
J., 11(4):588–595, 1965
 - $\mu(\textbf{T}) = \frac{\mu_0}{1 + \alpha \left| \textbf{T}_{\delta} \right|^{n-1}}$
 - Fits experimental data for: 0.6% w/w carboxymethyl cellulose
(CMC) solution in water, poly(vynil chloride)
- **Glen**
 - J. W. Glen. The creep of polycrystalline ice. Proc. R. Soc. A-Math.
Phys. Eng. Sci., 228(1175):519–538, 1955
 - $\mu(\textbf{T}) = \alpha \left| \textbf{T}_{\delta} \right|^{n-1}$
 - Fits experimental data for: ice
- **Seely**
 - Gilbert R. Seely. Non-newtonian viscosity of polybutadiene
solutions. AIChE J., 10(1):56–60, 1964
 - $\mu(\textbf{T}) = \mu_\infty +  (\mu_0 - \mu_\infty) e^{-\frac{\left| \textbf{T}_{\delta} \right|}{\tau_0}}$
 - Fits experimental data for: polybutadiene solutions
- **Blatter**
 - Erin C. Pettit and Edwin D. Waddington. Ice flow at low deviatoric
stress. J. Glaciol., 49(166):359–369, 2003
H Blatter. Velocity and stress-fields in grounded glaciers – a simple
algorithm for including deviatoric stress gradients. J. Glaciol.,
41(138):333–344, 1995
 - $\mu(\textbf{T}) = \frac{A}{( \left| \textbf(T)_{\delta} \right|^2  + \tau_0^2)^{\frac{n-1}{2}}}$
 - Fits experimental data for: ice

# 3 Rate Type Models

## 3.1 Maxwell, Oldroys and Burger models
- **Maxwell**
 - J. Clerk Maxwell. On the dynamical theory of gases. Philos. Trans.
R. Soc., 157:49–88, 1867
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $\textbf{S} + \lambda_1 \dot{\textbf{S}} = 2 \mu \textbf{D} $
 - $\textbf{M} =def \frac{\textbf{M}}{dt} - \textbf{LM} - (\textbf{ML})^T$
- **Oldroyd-B**
 - J. G. Oldroyd. On the formulation of rheological equations of state.
Proc. R. Soc. A-Math. Phys. Eng. Sci., 200(1063):523–541, 1950
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $\textbf{S} + \lambda \dot{\textbf{S}} = \eta_1 \textbf{A}_1 + \eta_2 \dot{\textbf{A}}_1 $

- **Oldroyd 8-constants**
 - J. G. Oldroyd. On the formulation of rheological equations of state.
Proc. R. Soc. A-Math. Phys. Eng. Sci., 200(1063):523–541, 1950
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $\textbf{S} + \lambda_1 \dot{\textbf{S}} + \frac{\lambda_3}{2} (\textbf{DS} + \textbf{SD}) 
         + \frac{\lambda_5}{2} (Tr\textbf{S})\textbf{D} + \frac{\lambda_6}{2} (\textbf{S} \cdott \textbf{D}) \textbf{I} = 2 \mu \textbf{D} $
- **Burgers**
 - J. M. Burgers. Mechanical considerations – model systems –
phenomenological theories of relaxation and viscosity. In First
report on viscosity and plasticity, chapter 1, pages 5–67.
Nordemann Publishing, New York, 1939
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $\textbf{S} + \lambda_1 \dot{\textbf{S}} + \lambda_2 \ddot{\textbf{S}} 
 = \eta_1 \textbf{A}_1 + \eta_2 \dot{\textbf{A}}_1 $
 
## 3.2 Giesekus

- H. Giesekus. A simple constitutive equation for polymer fluids
based on the concept of deformation-dependent tensorial mobility.
J. Non-Newton. Fluid Mech., 11(1-2):69–109, 1982
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $\textbf{S} + \lambda_1 \dot{\textbf{S}} + \frac{\alpha \lambda_2}{\mu} \textbf{S}^2 = - \mu \textbf{D} $


## 3.3 Phan-Thien-Tanner

- N. Phan Thien. Non-linear network viscoelastic model. J. Rheol.,
22(3):259–283, 1978
- N. Phan Thien and Roger I. Tanner. A new constitutive equation
derived from network theory. J. Non-Newton. Fluid Mech.,
2(4):353–365, 1977
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $Y \textbf{S} + \lambda \dot{\textbf{S}} + \frac{\lambda \xi}{2} (\textbf{DS} + \textbf{SD}) = - \mu \textbf{D} $
 - $Y = e^{-\epsilon \lambda / \mu Tr(\textbf(S)}$


## 3.4 Johnson-Segalman

- M. W. Johnson and D. Segalman. A model for viscoelastic fluid
behavior which allows non-affine deformation. J. Non-Newton.
Fluid Mech., 2(3):255–270, 1977
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $\textbf{S} = 2 \mu \textbf{D} + \textbf{S}'$
 - $\textbf{S}' + \left( \frac{d\textbf{S}'}{dt} + \textbf{S}' (\textbf{W}- a \textbf{D}) + (\textbf{W}- a \textbf{D})^T \textbf{S}' \right) = 2 \mu \textbf{D} $

## 3.5 Johnson-Tevaarwerk

- K. L. Johnson and J. L. Tevaarwerk. Shear behaviour of
elastohydrodynamic oil films. Proc. R. Soc. A-Math. Phys. Eng.
Sci., 356(1685):215–236, 1977
 - $\textbf{T} = \pi \textbf{I} + \textbf{S}$
 - $\dot{\textbf{S}} + \alpha sinh(\frac{\textbf{S}}{s_0}) = 2 \mu \textbf{D} $
- Fits experimental data for: lubricants


# 4 Integral type models

- B. Bernstein, E. A. Kearsley, and L. J. Zapas. A study of stress
relaxation with finite strain. Trans. Soc. Rheol., 7(1):391–410, 1963
I-Jen Chen and D. C. Bogue. Time-dependent stress in polymer
melts and review of viscoelastic theory. Trans. Soc. Rheol.,
16(1):59–78, 1972
- $T = \int \frac{\partial W}{\partial I} \textbf{C} \frac{\partial W}{\partial II} \textbf{C}^{-1} d\xi$
- Fits experimental data for: polyisobutylene, vulcanised rubber

## Visualization of the possible viscosity curves for each model

```python
import scipy as sp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
from scipy.optimize import curve_fit
```

# **Ostwald–de Waele power law**

```python
n = np.linspace(0.1,0.9,4)
mu0 = 250

absD = np.linspace(0.1,10, 1000)

df_n = pd.DataFrame(data=absD, columns=['absolutValueDeformationTensor'])
for n in n:
    res = mu0 * absD**(n-1)
    df_n[n] = res
df_n.plot(x='absolutValueDeformationTensor', legend=True)
```

# **Carreau Carreau–Yasuda**

## **Carreau Carreau–Yasuda** for fixed alpha = 1, 0 < n < 0.5:

```python
n = np.linspace(0,0.5,8)
mu0 = 250
mu_inf = 34

absD = np.linspace(0.000001,100, 1000)
df_n = pd.DataFrame(data=absD, columns=['absolutValueDeformationTensor'])

# varying n, alpha fixed = 1
for i in n:
    alpha = 1
    df_n[i] = mu_inf + (mu0 - mu_inf) / (1 + alpha * absD**2)**(i/2)
df_n.plot(x='absolutValueDeformationTensor', legend=True)
```

## **Carreau Carreau–Yasuda** for fixed n = 0.1, 0 < alpha < 4:

```python
df_alpha = pd.DataFrame(data=absD, columns=['absolutValueDeformationTensor'])
alpha_var = np.linspace(0,4,6)

# varying alpha, alpha fixed = 1
for i in alpha_var:
    n_fix = 0.1
    df_alpha[i] = mu_inf + (mu0 - mu_inf) / (1 + i * absD**2)**(n_fix/2)
df_alpha.plot(x='absolutValueDeformationTensor', legend=True)
```

## Second form of the equation:
## n<1, alpha = 2, a < 1

```python
n = np.linspace(0,0.5,5)
mu0 = 250
mu_inf = 72

absD = np.linspace(0.0001,10, 1000)
df_n = pd.DataFrame(data=absD, columns=['absolutValueDeformationTensor'])

# varying n, alpha fixed = 1
for n in n:
    alpha = 0.5
    a = 0.5
    res = mu_inf + (mu0 - mu_inf) * (1 + alpha * absD**a)**((n-1)/a)
    df_n[n] = res
df_n.plot(x='absolutValueDeformationTensor', legend=True)
```
