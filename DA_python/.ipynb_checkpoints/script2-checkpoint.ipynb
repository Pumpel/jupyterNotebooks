```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

 Read in data

```python
#data = np.loadtxt('rina2_c2/finished/7_Alte/Cond04/V1/postProcessing/forces/0/forces_clean.dat', skiprows=3)
data = np.loadtxt('forces.dat', skiprows=3)
```

Start to plot and do the magic starting from this value in the input data

```python
start=len(data)/2
#start=5000
```

Plot data

```python
fig = plt.figure()
fig.set_size_inches(12,6)

#plt.plot(data[start:,0],data[start:,1]+data[start:,4])
plt.plot(data[1000:,0],data[1000:,1]+data[1000:,4])

#plt.xlim(60,120)
plt.ylim(-60,-20)

plt.grid()
plt.show()
```

Create pandas dataFrame of data

```python
df=pd.DataFrame(data)
```

Compute rolling average of data

```python
#ndf=pd.rolling_mean(df,100,center=True)
#ndf=pd.rolling_mean(df,100)
ndf=df[start:].rolling(window=100,center=True).mean()
```

Extract x and y from ndf and drop NA

```python
df1=ndf[0].dropna()
#df1=ndf[0]

df2=ndf[1].dropna()+ndf[4].dropna()
#df2=ndf[1].dropna()+ndf[4]
```

Plot rolling averages and original data

```python
fig = plt.figure()
fig.set_size_inches(12,6)

plt.plot(data[start:,0],data[start:,1]+data[start:,4],c='blue',alpha=0.4)
plt.plot(df1,df2,c='red')

#plt.xlim(10,120)
#plt.ylim(-12.7,-12.4)

plt.grid()
plt.show()
```

Create vector for the interpolation function.
Vector starts from last time and middle time in the input data

```python
#subdivision=len(df1)
subdivision=20000

#xnew=np.linspace(60,119,60000)
xnew=np.linspace(int(data[-1][0]/2),int(data[-1][0]),subdivision)

#Uniform time step
dt=(data[-1][0]-data[-1][0]/2)/subdivision
```

Do the linear interpolation using the rolling mean

```python
from scipy import interpolate
fint = interpolate.interp1d(df1,df2,kind='linear')
ynew=fint(xnew-1)
```

Plot interpolated function

```python
fig = plt.figure()
fig.set_size_inches(12,12)

plt.plot(xnew, ynew,c='red')
#plt.plot(ndf[0],ndf[1]+ndf[4])
#plt.xlim(80,100)
#plt.ylim(-38,-36)
plt.grid()
```

Compute frequency on the interpolated function

```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt

fig = plt.figure()
fig.set_size_inches(12,6)

#f1, Pxx_den1 = signal.periodogram(ynew, 1/0.001)
f1, Pxx_den1 = signal.periodogram(ynew, 1/dt)

plt.xlim([0, 1])
plt.ylim([-0.1, 0.3])

plt.xlabel('frequency [Hz]')
plt.ylabel('PSD')

#plt.plot(f, Pxx_den)
plt.plot(f1, Pxx_den1,c='red')

#plt.semilogx()
#plt.semilogy()

plt.grid()
plt.show()

#fig.savefig('fre.png', dpi=300)
```

Compute the period form the frequency

```python
simperiod=1/f1[np.argmax(Pxx_den1)]
```

Compute the cut time

```python
reverse_periods=5
tcut=int(simperiod)*reverse_periods
```

Compute mean from cut time to last time

```python
(np.mean(data[df[df[0] > data[-1][0]-tcut].index[0]:df[df[0] > data[-1][0]-tcut].index[-1],1] + \
         data[df[df[0] > data[-1][0]-tcut].index[0]:df[df[0] > data[-1][0]-tcut].index[-1],4]))*(-2.)
```

```python
(np.std(data[df[df[0] > data[-1][0]-tcut].index[0]:df[df[0] > data[-1][0]-tcut].index[-1],1] + \
        data[df[df[0] > data[-1][0]-tcut].index[0]:df[df[0] > data[-1][0]-tcut].index[-1],4]))
```

```python
(np.var(data[df[df[0] > data[-1][0]-tcut].index[0]:df[df[0] > data[-1][0]-tcut].index[-1],1] + \
        data[df[df[0] > data[-1][0]-tcut].index[0]:df[df[0] > data[-1][0]-tcut].index[-1],4]))
```
