### Linear and polynomial regression

```python
import numpy as np
import statsmodels.api as sm
from sklearn import datasets, linear_model
from scipy import stats
```

```python
X = pdfile1[pdfile1.columns[0]]
Y = pdfile1[pdfile1.columns[1]]
```

```python
#compute linear regression using scipy
slope, intercept, r_value, p_value, std_err = stats.linregress(X,Y)

print "slope", slope, "intercept", intercept, "r:", r_value, "r-squared:", r_value**2
```

```python
# Get the linear models using numpy polyfit
coeffs = np.polyfit(X, Y, 1)
print "Polynomial coefficients", coeffs
```

### Box plot

```python
import matplotlib.pyplot as plt  

fig=plt.figure(figsize=(15, 15))

#n1=9;
xl=len(pdfile1.columns)
xc=4
yc=4

#meanlineprops = dict(linestyle='--', linewidth=2.5, color='purple')
meanlineprops = {'linestyle':'--', 'linewidth':1, 'color':'black'}

boxprops = dict(linestyle='--', linewidth=3, color='darkgoldenrod')
flierprops = dict(marker='o', markerfacecolor='green', markersize=3,linestyle='none',alpha=1.0/1.0)
medianprops = dict(linestyle='-.', linewidth=2.5, color='firebrick')
meanpointprops = dict(marker='D', markeredgecolor='black',markerfacecolor='firebrick')


for x in range(xl):
    fig.add_subplot(xc,yc,x+1)
    #pdfile3[pdfile3.columns[x]].plot(kind='box')
    pdfile1[pdfile1.columns[x]].plot(kind='box',grid=False,flierprops=flierprops,
                                     showmeans=True,meanline=True,meanprops=meanlineprops)
```

```python
#Boxplot of all columns with seaborn.
#In this case we normalize the dataframe using scikit learn.

from sklearn import preprocessing
import seaborn as sns

#sns.set_style("white")
#sns.set_style("dark")
#sns.set_style("whitegrid")
#sns.set_style("darkgrid")

#returns a numpy array
x = pdfile1.values 

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)

# Run the normalizer on the dataframe
#dfs = pd.DataFrame(x_scaled)
dfs = pd.DataFrame(x_scaled,columns=pdfile1.columns)

#sns.set_style("white")
#sns.set_style("dark")
#sns.set_style("whitegrid")
sns.set_style("darkgrid")

flierprops = dict(marker='o', markerfacecolor='red', markersize=6,linestyle='none',alpha=1.0/1.0)

fig=plt.figure(figsize=(20, 6))
ax = sns.boxplot(data=dfs,flierprops=flierprops)

#plt.savefig("bp1.pdf")


# reset rc params to defaults
sns.reset_orig()
```

### Scatter matrix plot

```python
import matplotlib.pyplot as plt   
import seaborn as sns
import matplotlib.patches as patches

#sns.set_style("white")
sns.set_style("whitegrid")
#sns.set_style("dark")
#sns.set_style("darkgrid")

fig=plt.figure(figsize=(30, 30))

N = len(pdfile1.columns)

for i in range(N):
#for i in range(2,N):
#for i in range(12,N):
    for j in range(N): 
        ax = fig.add_subplot(N,N,i*N+j+1) 

        plt.xlabel('', fontsize=0)
        plt.ylabel('', fontsize=0)
        
        #SCATTER AND REGRESSION PLOTS
        if i > j: 
            #ax.scatter(pdfile1[pdfile1.columns[j]], pdfile1[pdfile1.columns[i]], s=10, 
            #           alpha=1.0/2.0, color='steelblue')#, edgecolor="black")
            
            dfactor=2
            
            xd=pdfile1[pdfile1.columns[j]].std()/dfactor
            yd=pdfile1[pdfile1.columns[i]].std()/dfactor

            xmin=pdfile1[pdfile1.columns[j]].min()-xd
            xmax=pdfile1[pdfile1.columns[j]].max()+xd
            ymin=pdfile1[pdfile1.columns[i]].min()-yd
            ymax=pdfile1[pdfile1.columns[i]].max()+yd
            
            plt.xlim([xmin, xmax])
            plt.ylim([ymin, ymax])

            g = sns.regplot(pdfile1[pdfile1.columns[j]], pdfile1[pdfile1.columns[i]],
                        order=1,color="green",
                        scatter_kws={"s": 100,"alpha":0.25,"edgecolor":"black","linewidth":0},
                        line_kws={"color": "red","linewidth":2})
            
            ax.patch.set_facecolor('#F5F2EB')
            
            #NO AXIS NUMBERS OPTION1 - DO NOT RENDER OUTERMOST LABELS
            #ax.axes.get_xaxis().set_visible(False)
            #ax.axes.get_yaxis().set_visible(False)
            
            #NO AXIS NUMBERS OPTION2
            ax.axes.get_xaxis().set_ticks([])
            ax.axes.get_yaxis().set_ticks([])

            
        #EMPTY SQUARES OFF DIAGONAL OR SQUARES WITH CORRELATION TEXT
        elif i < j: 

            #Hide frame
            #ax.axis('off')

            #Hide axes and ticks
            #ax.axes.get_xaxis().set_ticks([])
            #ax.axes.get_yaxis().set_ticks([])
            ax.axes.get_xaxis().set_visible(False)
            ax.axes.get_yaxis().set_visible(False)
        
            ax.patch.set_facecolor('#F5F2EB')

            #Add text

            #ax.text(0.30,0.5,'%.4f' % corr.iloc[j,i], fontsize=40)
            ax.text(0.5, 0.5,
                    #"joel \n" '%.4f' % corr.iloc[j,i], fontsize=40,
                    '%.4f' % corr.iloc[j,i], size=50,
                    horizontalalignment='center',
                    verticalalignment='center',
                    transform=ax.transAxes)

            #The use of transform=ax.transAxes throughout the code indicates
            #that the coordinates are given relative to the axes bounding box, with 0,0 
            #being the lower left of the axes and 1,1 the upper right.

            #ax.set_axis_bgcolor('blue')
            
            
        #LABEL OUTER MOST AXES AND PLOT HISTOGRAMS 
        if j == 0: 
            ax.set_ylabel(pdfile1.columns[i],size='20')
        #if i == 0: 
        if i == N-1: 
            ax.set_xlabel(pdfile1.columns[j],size='20')
        if i == j: 
            ax.hist(pdfile1[pdfile1.columns[i]], bins=20, 
                    #histtype='stepfilled',
                    color='steelblue',linewidth=1,edgecolor='black') 
            ax.patch.set_facecolor('#F5F2EB')
            
            #NO AXIS NUMBERS OPTION1 - DO NOT RENDER OUTERMOST LABELS
            #ax.axes.get_xaxis().set_visible(False)
            #ax.axes.get_yaxis().set_visible(False) 
            
            #NO AXIS NUMBERS OPTION2
            ax.axes.get_xaxis().set_ticks([])
            ax.axes.get_yaxis().set_ticks([])
              
        #else: 
        #    #ax.scatter(data[j], data[i]) 
        #    ax.scatter(pdfile1[pdfile1.columns[j]], pdfile1[pdfile1.columns[i]])
        
        
#fig.subplots_adjust(hspace=0.2, wspace=0.2, left=0.08, bottom=0.08, top=0.9, right=0.9 )                
fig.subplots_adjust(hspace=0.25, wspace=0.25) 


#SET THE WHOLE FIGURE TO THE DESIRED BACKGROUND COLOR
fig.patch.set_facecolor('#F5F2EB')

fig.tight_layout()

plt.show() 


#SAVE BACKGROUND COLOR IN OUTPUT FILE
#fig.savefig('f2.png', facecolor='#F5F2EB')  
#fig.savefig('f2.pdf', facecolor='#F5F2EB')    
```
